#1.1 머신러닝이란?

일반적인 정의: 명시적인 프로그래밍 없이 컴퓨터가 학습하는 능력을 갖추게 하는 연구 분야이다.  (아서 새뮤얼)

공학적인 정의: 어떤 작업 T에 대한 컴퓨터 프로그램의 성능 P를 측정했을 때 경험 E로 인해 성능이 향상됐다면, 이 컴퓨터 프로그램은 작업 T와 성능 측정 P에 대해 경험 E로 학습한 것이다. (톰 미첼)







- training set: 시스템이 학습하는 데 사용되는 샘플

- training instance: 각 훈련 데이터

- accuracy: 정확히 분류된 데이터의 비율.



#1.2 왜 머신러닝을 사용하는가?

전통적인 프로그래밍 기법으로 스팸 필터를 만든다면 규칙이 점점 길고 복잡해지므로 유지보수를 하기가 힘들어짐.

하지만 머신러닝 기법에 기반을 둔 스팸 필터는 스팸 메일을 판단하는 기준을 자동으로 학습함. 따라서 프로그램이 짧아지고 유지보수가 쉬어짐. 정확도가 높아짐.

음성인식 또한 머신러닝이 유용한 또 다른 분야. 각 철자 피치의 사운드 강도를 측정하는 알고리즘보다는 각 단어를 녹음한 샘플을 사용해 스스로 학습하는 알고리즘이 더 좋은 솔루션.

데이터 마이닝: 머신러닝 기술을 적용해서 대용량의 데이터를 분석하여 겉으로는 보이지 않던 패턴을 발견하는 기법. 즉 이 기법을 통해 알고리즘이 학습한 것을 조사 가능함.















###뛰어난 분야 요약



*   기존 솔루션으로는 많은 수동 조정과 규칙이 필요한 문제: 머신러닝을 통해 코드를 간단하게 만듦 
*   전통적인 방식으로는 해결방법이 없는 복잡한 문제

*   유동적인 환경
*   복잡한 문제와 대용량 데이터에서 통찰 얻기(데이터 마이닝)


#1.4 머신러닝 시스템의 종류

넓은 범주로 분류하면



*   사람의 감독하에 훈련하는 것인지 그렇지 않은 것인지 : 지도, 비지도, 준지도, 강화 학습
*   실시간으로 점진적인 학습을 하는지 아닌지: 온라인 학습, 배치 학습
*   단순하게 알고 있는 데이터 포인트와 새 데이터 포인트를 비교하는 것인지 or 데훈련 데이터셋에서 패턴을 발견하여 예측 모델을 만드는지: 사례 기반 학습, 모델 기반 학습

##1.4.1 지도 학습과 비지도 학습

학습하는 동안의 감독 형태나 정보량에 따라 분류.

###1) 지도 학습

훈련 데이터에 레이블이 포함됨

    <종류>
    * 분류 (스팸 필터)
    * 회귀: feature을 사용하여 target 수치를 예측하는 것
        - 로지스틱 회귀: 클래스에 속할 확률 출력 (분류에 사용되는 회귀 알고리즘)










###2) 비지도 학습

훈련 데이터에 레이블 미포함

    <종류>
    * 계층 군집: 각 그룹을 더 작은 그룹으로 세분화 시킴
    * 시각화: 대규모의 고차원 데이터를 2D or 3D로 표현. 
      가능한 구조를 그대로 유지하려 하여 데이터가 어떻게 조직되어있는지 이해 가능. 예상하지 못한 패턴 발견 가능.
    * 차원 축소: 정보를 잃지 않으면서 데이터를 간소화 -> 상관관계가 있는 여러 특성을 하나로 합침(특성 추출)
    * 이상치 탐지: 학습 알고리즘에 주입하기 전에 데이터 셋에서 이상한 값을 자동으로 제거
    * 특이치 탐지: 훈련 세트에 있는 모든 샘플과 달라 보이는 새로운 샘플을 탐지
    * 연관 규칙 학습: 대량의 데이터에서 특성 간의 흥미로운 관계를 찾음
        (ex; 바비큐 소스와 감자를 구매한 사람이 스테이크도 구매하는 경향이 있다. 서로 가까이 진열하자)




###3) 준지도 학습

일부만 레이블이 있는 데이터를 다룸(지도 학습과 비지도 학습의 조함)

    * 심층 신뢰 신경망(DBN; deep belif network): 여러 겹으로 쌓은 (비지도 학습인)제한된 볼츠만 머신(RBM; restricted Bolzmann machine)에 기초함. 
      RBM이 비지도 학습 방식으로 훈련된 다음 전체 시스템이 지도 학습 방식으로 조정됨




###4) 강화 학습

환경을 학습해서 행동을 실행하고 그 결과로 reward(또는 penalty)를 받는 형식의 알고리즘

    
    * 에이전트: 학습하는 시스템
    * 정책(policy): 주어진 상황에서 에이전트가 어떤 행동을 선택해야 할지 정의
     -> 시간이 지나면서 가장 큰 보상을 얻기위해 최상의 전략(policy)를 스스로 학습

##1.4.2 배치 학습과 온라인 학습

###1) 배치 학습
시스템이 점진적으로 학습할 수 없는 기법.

- 시스템을 훈련시킨 후 제품에 적용하면 더 이상의 학습없이 실행. 즉, 학습한 것을 단지 적용만 함.

- 새로운 데이터에 대해 학습하려면 전체 데이터를 사용하여 시스템의 새로운 버전을 처음부터 다시 훈련해야 함. 그 후 이전 시스템을 중지시키고 새 시스템으로 교체해야 함. 

 => 시스템이 빠르게 변하는 데이터에 적응해야 한다면 더 능동적인 방법이 필요 (점진적인 학습 알고리즘 필요)





###2) 온라인 학습
- 데이터를 순차적으로 한개씩 또는 미니배치 단위로 주입하여 시스템 훈련.

- 학습 단계가 빠르고 비용이 적게 듦. 

- 데이터가 도착하는 대로 즉시 학습 가능.

    ※ learning rate: 온라인 학습에서 중요한 파라미터. 변화하는 데이터에 얼마나 빠르게 적응할 것인지를 나타냄. 

    - learning rate가 높을 시 -> 시스템이 데이터에 빠르게 적응하지만 예전 데이터를 금방 잊어버림
    - learning rate가 낮을 시 -> 시스템의 관성이 더 커져 더 느리게 학습되지만 새로운 데이터의 잡음 or 대표성 없는 데이터 포인트에 덜 민감해짐

- 온라인 학습의 단점: 시스템에 나쁜 데이터가 주입되었을 때 시스템 성능이 점진적으로 감소함

##1.4.3 사례 기반 학습과 모델 기반 학습

어떻게 일반화(generalize)되는가에 따라 머신러닝 분류 가능.

훈련 데이터로 학습하고 훈련 데이터에서는 본 적 없는 새로운 데이터에서 좋은 예측을 만들어야(일반화되어야) 함

-> 머신러닝의 진짜 목표: 새로운 샘플에 잘 작동하는 모델 만들기

###1) 사례 기반 학습

시스템이 훈련 샘플을 기억하여 유사도 측정을 사용해 새로운 데이터와 학습한 샘플을 비교하는 방식으로 일반화.

###2) 모델 기반 학습

샘플들의 모델을 만들어 예측에 사용

ex) 돈이 사람을 행복하게 만드는가? ( 나라별 1인당 GDP에 따른 삶의 만족도)

-> 1인당 GDP가 증가할수록 선형으로 삶의 만족도 증가

KakaoTalk_20201227_213306501 (1).jpg

이를 선형 함수로 모델링하면 (모델 선택)

-> 삶의 만족도 = θ_0 + θ_1 * 1인당 GDP (선형 모델)

->이 모델의 모델 파라미터: θ_0, θ_1

---

모델의 측정 지표

* 효용 함수(utility function) or 적합도 함수(fitness function): 모델이 얼마나 좋은지 측정
* 비용 함수(cost function): 모델이 얼마나 나쁜지 측정

---

<선형회귀 알고리즘>

알고리즘에 훈련 데이터를 공급하면 데이터에 가장 잘 맞는 선형 모델의 파라미터를 찾음 

-> 이를 훈련 시킨다 함




   







###지금까지의 작업 요약


1.   데이터를 분석한다.
2.   모델을 선택한다.
3.   훈련 데이터로 모델을 훈련시킨다. (학습 알고리즘이 비용 함수를 최소화하는 모델 파라미터를 찾는다.)
4.   새로운 데이터에 모델을 적용해 예측을 하고(추론), 이 모델이 잘 일반화되길 기대

##1.5.2 대표성 없는 훈련 데이터

일반화가 잘 되려면 우리가 일반화하기 원하는 새로운 사례를 훈련 데이터가 잘 대표하는 것이 중요함

* 샘플이 작으면 샘플링 잡음(sampling noise, 우연에 의한 대표성 없는 데이터) 발생
* 샘플이 매우 크다해도 표본 추출 방법이 잘못되면 대표성을 띠지 못할 수 있음
  
  -> The Literaray Digest의 샘플링 방식: 240만 명의 데이터 but 부유한 계층에 편중된 데이터 & 정치에 관심이 없는 사람 등을 제외하여 비응답 편향이 발생.
